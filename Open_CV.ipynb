## Face Detection

```python
img = cv2.imread('/content/humans.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
haar_cas = '/content/haarcascade_frontalface_alt.xml'
haar_cas
```

```python
# reading the XML file

face_cas = cv2.CascadeClassifier(haar_cas)
face_cas
```

```python
# detect the face

faces = face_cas.detectMultiScale(img,scaleFactor= 1.1,minNeighbors=5)
faces
```

```python
print('{} faces detected!!'.format(len(faces)))
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,200,0),5)
  plt.imshow(img)
```

# Number plate detection

```python
img = cv2.imread('/content/number_plate.JPG')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
har_cas = '/content/haarcascade_russian_plate_number.xml'
har_cas
```

```python
number_cas = cv2.CascadeClassifier(har_cas)
number_cas
```

```python
number = number_cas.detectMultiScale(img,scaleFactor= 1.1,minNeighbors=5)
number
```

```python
print('{} number plate detected!!'.format(len(number)))
for (x,y,w,h) in number:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,180),5)
  plt.imshow(img)
```
